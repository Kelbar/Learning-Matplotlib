```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'g')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Title')
#When I don't use Jupyter Notebook I have to user plt.show()
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'p')

plt.subplot(1,2,2)
plt.plot(x,y,'r')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y,'pr')
axes.set_xlabel('Tucan')
axes.set_ylabel('Pollo')
axes.set_title('Aves')
```

```python
# Plot inside plot
fig2 = plt.figure()

axes1 = fig2.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig2.add_axes([0.15,0.5,0.3,0.3])

axes1.plot(x,y,'g')
axes1.set_title('OUTSIDE PLOT')

axes2.plot(y,x,'r')
axes2.set_title('INSIDE PLOT')
```

```python
fig3, axes3 = plt.subplots(nrows=1,ncols=4)

for corr_ax in axes3:
    corr_ax.plot(x,y)

plt.tight_layout()
```

```python
fig4, axes4 = plt.subplots(nrows=1,ncols=2)

axes4[0].plot(x,y)
axes4[0].set_title('Title 1')

axes4[1].plot(y,x)
axes4[1].set_title('Title 2')

plt.tight_layout()
```

```python
fig4.savefig('Dual.png',dpi=200)
```

```python
fig5, axes5 = plt.subplots(nrows=2, ncols=1, figsize=(8,3))

axes5[0].plot(x,y)

axes5[1].plot(x*x,y,'r')

plt.tight_layout()
```

```python
fig6 = plt.figure()

axes6 = fig6.add_axes([0.1,0.1,1,1])

axes6.plot(x,y**3,'pr', label='CUBED')
axes6.plot(x,y**2,label='SQUARED')

plt.legend(loc=0)
```

**Appearance**

```python
fig7 = plt.figure()

axes7 = fig7.add_axes([0.1,0.1,0.8,0.8])


axes7.plot(x,y, color='red', lw=3, ls=':', alpha=0.8, marker='o', markersize=15, markeredgewidth=5, 
           markerfacecolor='yellow', markeredgecolor='blue')

axes7.set_xlim(0,1)
axes7.set_ylim(0,2)
```

```python
fig8, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options -, , -., :, steps
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```
